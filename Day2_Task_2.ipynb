```python
from PIL import Image
import numpy as np
import os

image_folder = "/content/archive/images"

image_arrays = []

for filename in os.listdir(image_folder):
    if filename.endswith((".png", ".jpg", ".jpeg")):

        img = Image.open(os.path.join(image_folder, filename))
        img = img.resize((150,150))
        img_array = np.array(img)

        image_arrays.append(img_array)

print("Total images:", len(image_arrays))
print("Shape of first image:", image_arrays[0].shape)
```

```python
from PIL import Image
import numpy as np
import os

image_folder = "/content/archive/images"

image_arrays = []

for filename in os.listdir(image_folder):
    if filename.endswith((".png", ".jpg", ".jpeg")):

        img = Image.open(os.path.join(image_folder, filename))

        img = img.resize((150, 150))

        img_array = np.array(img) / 255.0

        image_arrays.append(img_array)

print("Total images:", len(image_arrays))
print("Shape of first image:", image_arrays[0].shape)
print(image_arrays[0])
```

```python
from PIL import Image
import numpy as np
import os
from cv2 import cvtColor, COLOR_RGB2GRAY

image_folder = "/content/archive/images"

gray_images = []

for filename in os.listdir(image_folder):
    if filename.endswith((".png", ".jpg", ".jpeg")):

        img = Image.open(os.path.join(image_folder, filename))

        img = img.resize((150, 150))

        img_array = np.array(img)

        img_gray = cvtColor(img_array, COLOR_RGB2GRAY)

        gray_images.append(img_gray)

print("Total grayscale images:", len(gray_images))
print(gray_images[0])
```

```python
from PIL import Image, ImageOps
import matplotlib.pyplot as plt
import os

image_folder = "/content/archive/images"

files = os.listdir(image_folder)

for filename in files[:5]:
    if filename.endswith((".png", ".jpg", ".jpeg")):

        img = Image.open(os.path.join(image_folder, filename))
        flipped = ImageOps.mirror(img)

        plt.imshow(flipped)
        plt.title(filename)
        plt.axis("off")
        plt.show()
```

```python
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt
import os

image_folder = "/content/archive/images"

files = os.listdir(image_folder)

for filename in files[:5]:
    if filename.endswith((".png", ".jpg", ".jpeg")):

        img = Image.open(os.path.join(image_folder, filename))

        enhancer = ImageEnhance.Brightness(img)
        brighter = enhancer.enhance(1.5)

        plt.imshow(brighter)
        plt.title(filename)
        plt.axis("off")
        plt.show()
```

```python
from PIL import Image, ImageEnhance
import matplotlib.pyplot as plt
import os

image_folder = "/content/archive/images"

files = os.listdir(image_folder)

for filename in files[:5]:
    if filename.endswith((".png", ".jpg", ".jpeg")):

        img = Image.open(os.path.join(image_folder, filename))

        contrast = ImageEnhance.Contrast(img).enhance(2.0)

        plt.imshow(contrast)
        plt.title(filename)
        plt.axis("off")
        plt.show()
```
